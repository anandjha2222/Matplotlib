# Introduction of Matplotlib

Matplotlib is not a monolithic package, it is made up of below major modules:

**matplotlib.pyplot** : Higher level abstraction which works on data directly. These are the APIs that we use most often. We will use pyplot to setup the data that we want to visualize and then use the lower level APIs to customize our graph.

**pylab** : It's a convenience module which imports portions of the Matplotlib library and Numpy to give users a Matlab-like access to functionality available in Matplotlib

**Object level APIs (Matplotlib APIs)** : Core building blocks of our visualization, these allows us to exercise very fine grained control over how the data is displayed.

<h3>Matplotlib Object Hierarchy</h3>

Figure  ---> Axes  --->  Axis  ---> Major/Minor Ticks

Figure is a top level container. A Figure can be subduvided into Grids. Grids allow us to see where a particular data point is located at. 

PyPlot APIs operates at higher levels

Matplotlib APIs operates at lower levels

**Figure** : Overall window or page withing which all operations are performed, inside Figure there is Axes.

**Axes** : Area within a figure where actual graphs are plotted. Every Axes is made of 2 components, X-Axis and Y-Axis

Axis contains ticks (Major or Minor) , tick location, Labels

A Figure can have more than one Axes.

Imp Terms : Figure,Axes,Grid,Axis,Markers,Ticks,Legend

<h3>Figures and Axes</h3>

The Figure object represents the plotted figure.
Figure can be instantiated using *pyplot.figure()* function.

Then you can draw axes within the figure using *add_axes()* method of Figure object

Parameter to add_figure() is in the form of list like [0,0,1,1]
    1. 1st arg : This represent where in the figure the **left of the axes** will lie. 0 is the extreme left of the figure, 1 is the extreme right
    2. 2nd arg : This represents the **bottom location** of the axes in the figure. 0 is the bottom of the figure, 1 is the top
    3. 3rd arg : This is the **width** of the axes in terms of the fraction of the figure width
    4. 4th arg : This is the **height** of the axes in terms of the fraction of the figure height

<h4>Figure with only one axes</h4>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
```

<h4>Figure with multiple axes</h4>

```python
fig   = plt.figure()
ax1 = fig.add_axes([0,0.6,1,0.4])
ax2 = fig.add_axes([0.4,0,0.8,0.4])
```

```python
x = np.linspace(0,10,50)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0.6,1,0.4])
ax2 = fig.add_axes([0,0,0.8,0.4])

ax1.plot(x, np.sin(x))
ax1.set_xlabel('x',fontsize=15, color='r')
ax1.set_ylabel('sin(x)',fontsize=15, color='r')

ax2.plot(x, np.cos(x))
ax2.set_xlabel('x',fontsize=15, color='r')
ax2.set_ylabel('cos(x)',fontsize=15, color='r')
```

<h3>Alternate way to specify multiple Axes</h3>

Another way to specify axes within a matplotlib figure, but this time laid out in a matrix format, is to use the subplot functionality. The add_subplot() method of Figure class takes in a number for example 221 which basically means it will set up a 2 * 2 grid to our overall figure, and the subplot that we add will be to the first position in the grid. In a 2 * 2 grid , the first position refers to the first cell at the top left of the figure. Because there is just one subplot in below figure, it's hard to see the 2 * 2 grid. Wehen we add more subplots, matrix grid into which the figure has been divided can be seen.

```python
fig = plt.figure(figsize=(4,4))
ax1 = fig.add_subplot(221)
```

```python
type(ax1)
# AxesSubplot is basically axes drawn withing figure
```

Let's now see multiple subplots within a figure using plt.add_subplot()

```python
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)
ax1.plot([1,2,3,4],[2,4,6,8])

x = np.linspace(0,10,50)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x), label="Sine Curve")


ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x), label="Cosine Curve")

ax4 = fig.add_subplot(224)
ax4.plot(x, np.tan(x), label="Tan Curve")
```

Alternative to add_subplot function of Figure is subplot2grid() function which takes 2 tuples. 1st is for grid dimention and 2nd for axes position. 

Use this function in case you want further customization of subplots.

```python
ax1 = plt.subplot2grid((2,3),(0,0))
ax1.plot(x, np.sin(x))
ax1.set_label('sine curve')

ax2 = plt.subplot2grid((2,3),(0,1))
ax2.plot(x, np.cos(x))
ax2.set_label('cosine curve')

ax3 = plt.subplot2grid((2,3),(0,2), rowspan=2)
ax3.plot([1,2,3,4],[2,4,6,8])
ax3.set_label('straight line')
ax3.yaxis.tick_right()

ax4 = plt.subplot2grid((2,3),(1,0), colspan=2)
ax4.plot(x, np.exp2(x))
ax4.set_label('exponential curve')
```

<h2>Applying above all to a real world data - Example</h2>

```python
import pandas as pd

stock_data = pd.read_csv('resource/stock_px.csv')
stock_data.columns = ['Date', 'AA', 'AAPL', 'GE', 'IBM', 'JNJ', 'MSFT', 'PEP', 'SPX','XOM']
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.head()
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.05,0.65,0.5,0.3])

ax1.plot(stock_data['Date'], stock_data['AAPL'], color='green')
ax1.set_title('AAPL vs IBM (inset)')

ax2.plot(stock_data['Date'], stock_data['IBM'], color='blue')
```

```python
fig = plt.figure(figsize=(10,6))

fig.suptitle('Stock price comparision', fontsize=20)

ax1 = fig.add_subplot(221)
ax1.set_title('MSFT')
ax1.plot(stock_data['Date'], stock_data['MSFT'],color='green')

ax2 = fig.add_subplot(222)
ax2.set_title('GE')
ax2.plot(stock_data['Date'], stock_data['GE'],color='purple')

ax3 = fig.add_subplot(223)
ax3.set_title('IBM')
ax3.plot(stock_data['Date'], stock_data['IBM'],color='magenta')

ax4 = fig.add_subplot(224)
ax4.set_title('SPX')
ax4.plot(stock_data['Date'], stock_data['SPX'],color='orange')
```
